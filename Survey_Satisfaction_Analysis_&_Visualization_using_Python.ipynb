**Step 1: Import Libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Step 2:Load Dataset**

```python
# Load CSV file
df = pd.read_csv("customer_satisfaction_data.csv")
```

**Step 3: Inspect Satisfaction Column**

```python
# Check unique satisfaction levels
df['Satisfaction Score'].unique()
# Count missing values (if any)
df['Satisfaction Score'].isnull().sum()
```

**Step 4: Count Each Satisfaction Level**

```python
# Count how many people selected each level
satisfaction_counts = df['Satisfaction Score'].value_counts()
print(satisfaction_counts)
```

**Step 5: Plot Bar Chart (Seaborn)**

```python
# Set plot style
sns.set(style="whitegrid")

# Create bar plot
plt.figure(figsize=(8,5))
sns.barplot(x=satisfaction_counts.index, y=satisfaction_counts.values, palette="viridis")
plt.title("Survey Satisfaction Level Counts")
plt.xlabel("Satisfaction Level")
plt.ylabel("Number of Respondents")
plt.xticks(rotation=45)
plt.show()

# Save as PNG
plt.savefig("satisfaction_bar_chart.png", dpi=300, bbox_inches='tight')
plt.show()
```

**Step 6: Sort Levels**

```python
order = ["Very Dissatisfied", "Dissatisfied", "Neutral", "Satisfied", "Very Satisfied"]

plt.figure(figsize=(8,5))
sns.barplot(x=satisfaction_counts.index, y=satisfaction_counts.values, order=order, palette="viridis")
plt.title("Survey Satisfaction Level Counts")
plt.xlabel("Satisfaction Level")
plt.ylabel("Number of Respondents")
plt.show()
```

**Step 7: Handle Missing Values**

```python
# Drop rows where Satisfaction is missing
df = df.dropna(subset=['Satisfaction Score'])

# Optional: confirm no missing values remain
print(df['Satisfaction Score'].isnull().sum())
```

**Step 8: Add Percentages on Bars**

```python
# Recalculate counts after dropping missing values
satisfaction_counts = df['Satisfaction Score'].value_counts()

plt.figure(figsize=(8,5))
sns.barplot(x=satisfaction_counts.index, y=satisfaction_counts.values, palette="viridis")

# Add percentages on top of each bar
total = satisfaction_counts.sum()
for i, count in enumerate(satisfaction_counts):
    percentage = (count / total) * 100
    plt.text(i, count + 1, f'{percentage:.1f}%', ha='center')

plt.title("Survey Satisfaction Level Counts")
plt.xlabel("Satisfaction Level")
plt.ylabel("Number of Respondents")
plt.xticks(rotation=45)
plt.show()
```
